```python
means = []
for col in df2.columns:
    means.append((df2[col].mean(), col))
means.sort(reverse=True)
n = 5
top_n = []
for mean, col in means[:n]:
    top_n.append(col)
print(top_n)
```

```python
means = []
for col in df2.columns:
    if col == 'Date':
        continue
    means.append((df2[col].mean(), col))
means.sort(reverse=True)
n = 10
top_n = []
for mean, col in means[:n]:
    top_n.append(col)
print(top_n)
```

```python
for col in top_n:
    plt.plot(df.Date.str[-4:], df[col])
plt.xticks(rotation=45)
plt.legend(top_n)
plt.ylabel('%')
ax = plt.gca()
temp = ax.xaxis.get_ticklabels()
plt.show()
```
